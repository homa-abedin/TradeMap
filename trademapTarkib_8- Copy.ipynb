### توابع مربوط به دریافت کد کالا و نام کشور


```python
# نام کالا رو میگیره و توی متغییر ذخیره می کنه

last_product_code = None  # متغیر برای نگهداری آخرین کد معتبر

def get_selected_product_code(driver):
    """
    دریافت کد محصول انتخاب‌شده از dropdown در سایت TradeMap
    """
    global last_product_code
    
    # پیدا کردن کادر انتخابی
    dropdown = Select(driver.find_element(By.XPATH, '//*[@id="ctl00_NavigationControl_DropDownList_Product"]'))
    
    # متن گزینه انتخاب شده
    selected_value = dropdown.first_selected_option.text.strip()
    
    # پیدا کردن اولین عدد (کد)
    match = re.search(r'\d+', selected_value)
    if match:
        code = match.group()
        if len(code) == 6:   # بررسی طول کد
            last_product_code = code   # مقدار جدید ذخیره میشه
    
    return last_product_code  # همیشه مقدار معتبر (آخرین ۶ رقمی) برمی‌گرده
    
#---- USE:--- current_code= get_selected_product_code(driver)
```

```python
def ensure_page(driver, expected_url):
    """
    بررسی می‌کند که در صفحه مورد انتظار هستیم. 
    اگر نه، به URL مورد نظر هدایت می‌شود.
    """
    current_url1 = driver.current_url
    if current_url1 != expected_url:
        print(f"در صفحه اشتباهی هستید: {current_url1}. بازگشت به صفحه اصلی...")
        driver.get(expected_url)
        time.sleep(2)  # صبر برای بارگذاری صفحه
    else:
        print("در صفحه مورد نظر هستید.")
```

```python
def filter_products_by_importers(df, product_col="Product code", importer_col="importer", threshold=254):
    """
    بر اساس کد محصول (۶ رقم اول) و تعداد واردکننده‌ها، فیلتر می‌کند.
    
    پارامترها:
        df (pd.DataFrame): دیتافریم اصلی
        product_col (str): نام ستون کد محصول
        importer_col (str): نام ستون واردکننده
        threshold (int): آستانه مقایسه
        
    خروجی:
        pd.DataFrame: شامل کد محصول (۶ رقم اول) و تعداد واردکننده‌ها
    """
    
    result = (
        df
        .groupby(df[product_col].astype(str).str[:6])[importer_col]
        .nunique()
        .reset_index()
        .rename(columns={product_col: "ProductCode6Digit", importer_col: "UniqueImporters"})
    )
    
    # فیلتر روی threshold
    filtered = result[result["UniqueImporters"] < threshold]
    
    return filtered
```
