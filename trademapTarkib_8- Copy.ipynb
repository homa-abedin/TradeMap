```python
df=pd.DataFrame()
finaldf1= pd.DataFrame()
num_rows = len(driver.find_elements(By.XPATH, '//*[@id="ctl00_PageContent_MyGridView1"]/tbody/tr[position()>1]'))

for i in range(1, num_rows + 1):
    # هر بار ردیف رو دوباره پیدا کن (i+1 چون tr[1] هدره)
    row = driver.find_element(By.XPATH, f'//*[@id="ctl00_PageContent_MyGridView1"]/tbody/tr[{i+1}]')
    
    product_code=row.find_element(By.XPATH, './td[2]/a').text.strip()
    cell = row.find_element(By.XPATH, './td[3]')

    # سعی کن title رو بگیری
    product_name = cell.get_attribute('title')

    # اگر title وجود نداشت، از متن استفاده کن
    if not product_name:
        product_name = cell.text.strip()
    print(f"Clicking product {i}: Code={product_code}, Name={product_name}")

    # مطمئن شو لینک وجود داره
    link = row.find_element(By.XPATH, './td[2]//a')
    # کلیک روی لینک
    link.click()
    time.sleep(2)  # یا waits تا صفحه لود بشه
    all_tables = pd.read_html(driver.page_source, attrs={'id': 'ctl00_PageContent_MyGridView1'})
    temp = all_tables[0]
    df = temp.iloc[:, 1:]
    df.reset_index(drop=True, inplace=True)
    fixed_columns = ['Product code', 'Product label','Exporters'] + [str(y) for y in range(2005, 2025)]
    df.columns = [str(col).split()[-1] if "Imported value in" in str(col) else str(col) for col in df.columns]
    df['Product code']=product_code
    df['Product label']=product_name
    df.columns
    # اضافه کردن ستون‌های گمشده 
    missing_cols = set(fixed_columns) - set(df.columns)
    for col in missing_cols:
        df[col] = 0
    # مرتب‌سازی ستون‌ها و اضافه کردن ستون Partner
    df = df[fixed_columns]
    df.drop(df[df['Exporters'] == 'World'].index, inplace=True)
    df=df.fillna(0)
    finaldf1=pd.concat([df, finaldf1], ignore_index=True)
    print(df)
    # برگشتن به صفحه قبلی (لیست کالاها)
    driver.back()
    time.sleep(2)
```
